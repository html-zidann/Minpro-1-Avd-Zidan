```python
results = []

cols = df.select_dtypes(include=['float64', 'int64'])

for col in cols:
  q1 = df[col].quantile(0.25)
  q3 = df[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5*iqr
  upper_bound = q3 + 1.5*iqr
  outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
  percent_outliers = (len(outliers)/len(df))*100
  results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

# Dataframe dari list hasil
results_df = pd.DataFrame(results)
results_df.set_index('Kolom', inplace=True)
results_df = results_df.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)

# Tampilkan dataframe
display(results_df)
```

```python
columns_to_impute = ["Case Number", "Date", "Year", "Type", "Country", "Area", "Location", "Activity", "Name", "Sex", "...", "Species", "Investigator or Source", "pdf", "href formula\t", "href", "Case Number.1", "Case Number.2", "original order", "Unnamed: 22", "Unnamed: 23"]

for col in columns_to_impute:
    # Check if the column exists in the DataFrame
    if col in df.columns:
        # Check if the column is numeric
        if pd.api.types.is_numeric_dtype(df[col]):
            Q1 = df[col].quantile(0.25)
            Q3 = df[col].quantile(0.75)
            IQR = Q3 - Q1
            lower_bound = Q1 - 1.5 * IQR
            upper_bound = Q3 + 1.5 * IQR

            # Using .loc[] to avoid SettingWithCopyWarning
            df.loc[:, col] = df[col].clip(lower=lower_bound, upper=upper_bound)
        else:
            print(f"Column '{col}' is not numeric. Skipping outlier handling...") # Print a message if the column is not numeric
    else:
        print(f"Column '{col}' not found in DataFrame. Skipping...") # Print a warning message if the column is not found
```

# **Visualisasi**

```python
country_counts = df['Country'].value_counts().head(10)
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x=country_counts.index, y=country_counts.values, color='purple')
plt.title('Distribusi Serangan Hiu Berdasarkan Negara (Top 10)')
plt.xlabel('Negara')
plt.ylabel('Jumlah Serangan')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
filtered_df = df[df['Year'] >= 1500]
year_fatal_counts = filtered_df.groupby('Year')['Fatal (Y/N)'].value_counts().unstack(fill_value=0)
plt.figure(figsize=(12, 6))
sns.scatterplot(x=year_fatal_counts.index, y=year_fatal_counts['Y'], color='purple', alpha=0.5)
plt.title('Hubungan antara Fatalitas dan Tahun (Mulai Tahun 1500)')
plt.xlabel('Tahun')
plt.ylabel('Jumlah Fatalitas')
plt.xticks(rotation=45, ha='right')
plt.xlim(1500, df['Year'].max())
plt.show()
```

```python
location_stats = df.groupby('Area').agg(
    serangan_per_negara=('Case Number', 'count'),
    persentase_serangan_fatal=('Fatal (Y/N)', lambda x: (x == 'Y').mean() * 100)
).reset_index()
top_locations = location_stats.nlargest(10, 'serangan_per_negara')
top_locations.set_index('Area')[['serangan_per_negara', 'persentase_serangan_fatal']].plot(kind='bar')
plt.title('Perbandingan Jumlah Serangan vs Rata-rata Tingkat Keparahan Cedera')
plt.ylabel('Jumlah / Tingkat Keparahan')
plt.xticks(rotation=45)
plt.show()
```

# **Insight Dari Visualisasi Data Serangan Hiu**

**1. Distribusi Serangan Hiu Berdasarkan Negara**

Sebagian besar serangan hiu terkonsentrasi di beberapa negara, seperti Amerika Serikat dan Australia. Hal ini menunjukkan bahwa negara-negara tersebut mempunyai perairan yang lebih sering dikunjungi hiu atau memiliki aktivitas manusia di laut yang lebih tinggi. Ada beberapa negara dengan jumlah serangan yang relatif rendah, menunjukkan risiko yang lebih kecil di wilayah tersebut.

**2. Hubungan antara Fatalitas dan Tahun**

Secara umum, ada tren peningkatan jumlah serangan hiu yang fatal seiring berjalannya waktu, khususnya sejak tahun 1900-an. Ini kemungkinan disebabkan oleh peningkatan aktivitas manusia di laut, perubahan iklim, atau faktor lainnya. Namun, perlu diingat bahwa data serangan hiu di masa lampau mungkin kurang lengkap.

**3. Perbandingan Jumlah Serangan dan Tingkat Keparahan di Berbagai Lokasi**

Beberapa lokasi memiliki jumlah serangan hiu yang tinggi dengan persentase fatalitas yang tinggi juga, menunjukkan bahwa area tersebut berisiko tinggi. Lokasi lain mungkin memiliki jumlah serangan yang tinggi tetapi persentase fatalitas yang rendah, menunjukkan bahwa serangan di area tersebut cenderung tidak fatal.
